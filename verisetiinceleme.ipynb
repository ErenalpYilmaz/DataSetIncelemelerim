```python
# 'United Kingdom' kelimesini içeren satırları seçelim
uk_filmleri = df[df['production_countries'].str.contains('United Kingdom', na=False)]
```

```python
uk_filmleri['genres'] = uk_filmleri['genres'].fillna('').str.split(',')
```

```python
# 3. Türlerin formatını normalize edelim (küçük harf ve boşlukları temizleyelim)
uk_filmleri['genres'] = uk_filmleri['genres'].apply(lambda x: [genre.strip().lower() for genre in x if genre != ''])
```

```python
uk_filmleri = uk_filmleri.explode('genres')
```

```python
# 5. Tekrarları önlemek için her film-tür kombinasyonunu bir kere ele alalım
uk_filmleri = uk_filmleri.drop_duplicates(subset=['movie_id', 'genres'])
```

```python
# Türlere göre ortalama revenue, vote_count ve vote_average değerlerini hesaplayalım
tur_bazli_siralama = uk_filmleri.groupby('genres').agg({
    'revenue': 'mean',
    'vote_count': 'mean',
    'vote_average': 'mean'
}).sort_values(by='vote_average', ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=tur_bazli_siralama.index, y=tur_bazli_siralama['vote_average'])
plt.title('Türlere Göre Ortalama Oy (Vote Average)')
plt.xlabel('Film Türleri')
plt.ylabel('Ortalama Oy')
plt.xticks(rotation=45)  # X eksenindeki etiketleri döndür
plt.tight_layout()  # Grafik dışına taşmaması için
plt.show()
```

# Hangi yapimcilar daha fazla film uretiyor ve bunlarin basari oranlari

Başarı Oranı: Her türün ortalama puanını ve o türün toplam gelirinin, şirketin toplam gelirine oranını kullanarak hesaplayabiliriz. Başarı oranını, her tür için ortalama puan ile o türün hasılatının toplam hasılata oranını çarparak bulabiliriz.

```python
# Dataseti oku
df = pd.read_csv('Movies.csv')

# Yapımcıların sayısını ve ortalama puanlarını hesaplayalım
production_analysis = df.groupby('production_companies').agg(
    film_sayisi=('title', 'count'),  # Film sayısı
    toplam_gelir=('revenue', 'sum'),  # Toplam gelir
    ortalama_puan=('vote_average', 'mean')  # Ortalama puan
).reset_index()

# Sonuçları sıralayalım
production_analysis = production_analysis.sort_values(by='film_sayisi', ascending=False)

# Her yapımcının başarı oranını hesaplayalım
production_analysis['basari_orani'] = production_analysis['ortalama_puan'] * (production_analysis['toplam_gelir'] / production_analysis['toplam_gelir'].sum())

# En çok film üreten ilk 10 yapımcıyı alalım
top_producers = production_analysis.head(10)

# Grafik boyutunu ayarlayalım
plt.figure(figsize=(12, 6))

# Yatay Bar Grafiği: Film Sayısı
plt.subplot(1, 2, 1)
plt.barh(top_producers['production_companies'], top_producers['film_sayisi'], color='skyblue')
plt.title('En Çok Film Üreten Yapımcılar', fontsize=16)
plt.xlabel('Film Sayısı', fontsize=14)
plt.ylabel('Yapımcılar', fontsize=14)
plt.gca().invert_yaxis()  # Yapımcıları yukarıdan aşağı sıralamak için

# Dikey Bar Grafiği: Başarı Oranı
plt.subplot(1, 2, 2)
plt.bar(top_producers['production_companies'], top_producers['basari_orani'], color='salmon')
plt.title('Yapımcıların Başarı Oranı', fontsize=16)
plt.xlabel('Yapımcılar', fontsize=14)
plt.ylabel('Başarı Oranı', fontsize=14)
plt.xticks(rotation=45, ha='right')  # X ekseninde etiketleri döndürme

# Grafiği göster
plt.tight_layout()
plt.show()
```

# Tablolar incelendiğinde Metro-Goldwyn-Mayer şirketinin hem en çok film ürettiği hemde en yüksek başarı oranlarına sahip şirketlerden birisi olduğu kanaatine vardım. Bizim için en önemli olan şey hem hız hemde başarı ve bunlar bu şirkette toplanmış. Bu verilerden dolayı şirketi biraz daha detaylı incelemeye karar verdim.

```python
df = pd.read_csv('Movies.csv')

# Paramount yapımcısına ait filmleri filtrele
paramount_films = df[df['production_companies'].str.contains('Paramount', na=False)].copy()

# Filmleri türlerine göre ayır
paramount_films['turler_listesi'] = paramount_films['genres'].apply(lambda x: list(set(x.split(','))) if pd.notna(x) else [])

# Her bir tür için ortalama puanı ve toplam geliri hesapla
tur_basari = paramount_films.explode('turler_listesi').groupby('turler_listesi').agg(
    ortalama_oy=('vote_average', 'mean'),
    toplam_gelir=('revenue', 'sum'),
    film_sayisi=('title', 'count')
).reset_index()

# Paramount yapımcısının toplam gelirini hesapla
toplam_gelir_paramount = paramount_films['revenue'].sum()

# Başarı oranını hesapla
tur_basari['basari_orani'] = tur_basari['ortalama_oy'] * (tur_basari['toplam_gelir'] / toplam_gelir_paramount)

# Başarı oranına göre sıralama
tur_basari_sirali = tur_basari.sort_values(by='basari_orani', ascending=False)

# Genel başarı oranını hesapla
genel_basari_orani = tur_basari['basari_orani'].mean()

# Sonuçları göster
print(tur_basari_sirali)
print("\nParamount Yapımcısının Genel Başarı Oranı:", genel_basari_orani)
```

# Bütçe ve gelir arasındaki ilişki

```python
df = pd.read_csv('Movies.csv')
# Sıralama işlemi (en yüksek izleyici puanına göre)
top_movies = df.nlargest(1000, 'vote_average')

top_movies = top_movies[top_movies['budget'] > 0]  # Bütçesi sıfırdan büyük olanları alalım
top_movies = top_movies[top_movies['revenue'] > 0]  # Geliri sıfırdan büyük olanları alalım
```

#En fazla bütçeye sahip olup da en az parayı kazanan filmlerin türlerini grafikte gösterimi

```python
df = pd.read_csv('Movies.csv')
highest_budget_movies = df.nlargest(1000000, 'budget')  # İlk 100 filmi seçelim

# En düşük gelir elde eden filmleri filtreleyelim
lowest_revenue_movies = highest_budget_movies[highest_budget_movies['revenue'] <= 1000000]  # Geliri 1,000,000 USD altında olanları seçelim

# Türleri ayıralım ve sayalım
genre_counts = lowest_revenue_movies['genres'].str.get_dummies(sep=', ').sum()

# Toplam film sayısını hesaplayalım
total_movies = lowest_revenue_movies.shape[0]

# Yüzde değerleri hesaplayalım
genre_percentages = (genre_counts / total_movies) * 100

# Türleri filtreleme: Yüzde değeri 0'dan büyük olanları seç
genre_percentages = genre_percentages[genre_percentages > 0]

# Bar grafiğini oluşturma
plt.figure(figsize=(10, 6))
bars = plt.bar(genre_percentages.index[:10], genre_percentages.values[:10], color='lightcoral')

# Değerleri grafik üzerinde gösterme
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.1f}%', ha='center', va='bottom')  # Yüzde değerlerini yazdır

# Grafiği özelleştirelim
plt.title('En Yüksek Bütçeye Sahip Olup En Az Gelir Elde Eden Filmlerin Türleri (%)', fontsize=16)
plt.ylabel('Film Yüzdesi (%)', fontsize=14)
plt.xlabel('Türler', fontsize=14)

# Grafik etiketlerini döndür
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('Movies.csv')
# En düşük bütçeye sahip filmleri seçelim
lowest_budget_movies = df.nsmallest(100000, 'budget')  # İlk 100 filmi seçelim

# Gelir/bütçe oranını hesaplayalım
lowest_budget_movies['profit_ratio'] = lowest_budget_movies['revenue'] / lowest_budget_movies['budget']

# Türleri ayıralım ve dummies oluşturup birleştirelim
genres_dummies = lowest_budget_movies['genres'].str.get_dummies(sep=', ')
lowest_budget_movies = lowest_budget_movies.join(genres_dummies)

# Ortalama kazanç oranını hesaplayalım
profit_ratios_by_genre = {}

for genre in genres_dummies.columns:
    genre_movies = lowest_budget_movies[lowest_budget_movies[genre] == 1]
    if not genre_movies.empty:
        average_profit_ratios = genre_movies['profit_ratio'].mean()
        profit_ratios_by_genre[genre] = average_profit_ratios

# DataFrame oluşturalım
average_profit_ratio_df = pd.DataFrame(profit_ratios_by_genre.items(), columns=['Genre', 'Average Profit Ratio'])

# Bar grafiğini oluşturma
plt.figure(figsize=(10, 6))
bars = plt.bar(average_profit_ratio_df['Genre'][:10], average_profit_ratio_df['Average Profit Ratio'][:10], color='lightgreen')

# Değerleri grafik üzerinde gösterme
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', ha='center', va='bottom')  # Oranı yazdır

# Grafiği özelleştirelim
plt.title('En Düşük Bütçeye Sahip Olup En Yüksek Gelir Elde Eden Filmlerin Türleri (Kazanç Oranı)', fontsize=16)
plt.ylabel('Kazanç Oranı (Gelir/Bütçe)', fontsize=14)
plt.xlabel('Türler', fontsize=14)

# Grafik etiketlerini döndür
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#**2008 yılında çıkan en başarılı filmler ve bu filmlerin yapımcıları**

```python
df = pd.read_csv('Movies.csv')
# Tarih formatını düzenleme (errors='coerce' ile hatalı tarihleri NaT yapar)
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# 2008 yılında çıkan filmleri seçelim (NaT olan tarihleri çıkar)
movies_2008 = df[df['release_date'].dt.year == 2008].dropna(subset=['release_date'])

# Başarı kriteri olarak en yüksek geliri kullanarak en başarılı ilk 20 filmi seçelim
top_20_movies = movies_2008.nlargest(20, 'revenue')

# Türler ve yapımcılar için verileri ayıklayalım
top_20_movies['genres'] = top_20_movies['genres'].str.split(',')  # Türleri ayır
top_20_movies['production_companies'] = top_20_movies['production_companies'].str.split(',')  # Yapımcıları ayır

# Türlerin sayısını ve yapımcıların sayısını hesaplayalım
genre_counts = top_20_movies.explode('genres')['genres'].value_counts()
production_counts = top_20_movies.explode('production_companies')['production_companies'].value_counts()

# Gelirleri alalım
revenue_data = top_20_movies[['title', 'revenue']]

# Grafik oluşturma için ayarlamalar
fig, axes = plt.subplots(2, 1, figsize=(12, 12))

# Türlerin grafik gösterimi
axes[0].bar(genre_counts.index[:10], genre_counts.values[:10], color='lightblue')
axes[0].set_title('2008 Yılında Çıkan En Başarılı Filmlerin Türleri', fontsize=16)
axes[0].set_ylabel('Film Sayısı', fontsize=14)
axes[0].set_xlabel('Türler', fontsize=14)
axes[0].tick_params(axis='x', rotation=45)

# Yapımcıların grafik gösterimi
axes[1].bar(production_counts.index[:10], production_counts.values[:10], color='lightgreen')
axes[1].set_title('2008 Yılında Çıkan En Başarılı Filmlerin Yapımcıları', fontsize=16)
axes[1].set_ylabel('Film Sayısı', fontsize=14)
axes[1].set_xlabel('Yapımcılar', fontsize=14)
axes[1].tick_params(axis='x', rotation=45)

# Grafiklerin ayarlanması
plt.tight_layout()
plt.show()

# Gelir verilerini gösterme
print("2008 Yılında Çıkan En Başarılı İlk 20 Film ve Gelirleri:")
print(revenue_data)
#

```

#**Hangi yapımcı daha fazla film üretiyor ve bu filmlerin başarı oranları ne ?**

```python
df = pd.read_csv('Movies.csv')
# 'production_companies' sütununda NaN olan değerleri boş string olarak değiştirin
df['production_companies'] = df['production_companies'].fillna('')

# Yapımcı şirketlerinin veri tipini string'e çevirin
df['production_companies'] = df['production_companies'].astype(str)

# Yapımcı şirketleri ayıralım (Birden fazla yapımcı olabilir)
df['production_companies'] = df['production_companies'].str.split(',')

# Veriyi genişletelim, her yapımcı için ayrı satır olacak şekilde
df_exploded = df.explode('production_companies')

# Oy sayısı en az 1000 olan filmleri filtreleyelim
df_exploded = df_exploded[df_exploded['vote_count'] >= 1000]

# Yapımcı şirketlerine göre gruplama ve ortalama puanların hesaplanması
producer_ratings = df_exploded.groupby('production_companies').agg(
    average_vote=('vote_average', 'mean'),
    total_votes=('vote_count', 'sum'),
    movie_count=('production_companies', 'count')
).reset_index()

# En başarılı yapımcıları seçelim (örneğin en çok puan alan ilk 10)
top_producers = producer_ratings.nlargest(20, 'average_vote')

# Grafik oluşturma
plt.figure(figsize=(10, 6))

# Bar grafiği oluşturma
bars = plt.bar(top_producers['production_companies'], top_producers['average_vote'], color='skyblue')

# Grafikte değerleri gösterme
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', ha='center', va='bottom')

# Grafik başlığı ve eksen etiketleri
plt.title('En Başarılı Yapımcılar (Ortalama Oy Puanına Göre)', fontsize=16)
plt.xlabel('Yapımcı Şirketler', fontsize=14)
plt.ylabel('Ortalama Oy Puanı', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Grafiği gösterme
plt.show()

# Ayrıca başarı kriterlerini detaylı incelemek için veri çıktısı
print("En Başarılı Yapımcılar ve Filmlerinin Ortalama Puanları:")
print(top_producers)
```
